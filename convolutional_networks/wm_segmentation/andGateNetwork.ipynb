## AND Gate network with backpropagation

```python
import math
import random
```

```python
b = 0.45
a = 0.9
```

```python
n00 = 1
n01 = 1
N20Ref = 1
```

```python
# initial values
w00 = 0.4
w01 = 0.1
w02 = -0.1
w03 = -0.1
w10 = 0.06
w11 = -0.4
w00_ant = w00
w01_ant = w01
w02_ant = w02
w03_ant = w03
w10_ant = w10
w11_ant = w11

pdW10 = 0
pdW11 = 0
pdW00 = 0
pdW01 = 0
pdW02 = 0
pdW03 = 0

n00 = 1
n01 = 1
# epochs
for i in range(100):
    #n00 = random.uniform(0, 1)
    #n01 = random.uniform(0, 1)
    n00 = 1
    n01 = 1
    # forward propagation
    x1 = w00*n00 + w01*n01
    x2 = w02*n00 + w03*n01
    N10 = 1/(1+math.exp(-x1))
    N11 = 1/(1+math.exp(-x2))
    x3 = w10*N10 + w11*N11
    N20 = 1/(1+math.exp(-x3))
    # backpropagation to the output layer
    N20Ref = 1 if n00==1 and n01==1 else 0
    N20E = N20*(1-N20)*(N20Ref-N20)
    dW10 = b*N20E*N10
    dW11 = b*N20E*N11
    w10 = w10 + dW10 + a*pdW10
    w11 = w11 + dW11 + a*pdW11
    #pdW10 = w10 - w10_ant
    #pdW11 = w11 - w11_ant
    w10_ant = w10
    w11_ant = w11
    # backpropagation to the hidden layer
    N10E = N20E*w10
    N11E = N20E*w11
    dW00 = b*N10E*n00
    dW01 = b*N10E*n01
    dW02 = b*N11E*n00
    dW03 = b*N11E*n01
    w00 = w00 + dW00 + a*pdW00
    w01 = w01 + dW01 + a*pdW01
    w02 = w02 + dW02 + a*pdW02
    w03 = w03 + dW03 + a*pdW03
    #pdW00 = w00 - w00_ant
    #pdW01 = w01 - w01_ant
    #pdW02 = w02 - w02_ant
    #pdW03 = w03 - w03_ant
    w00_ant = w00
    w01_ant = w01
    w02_ant = w02
    w03_ant = w03
    
    
    n00 = 1
    n01 = 0
    # forward propagation
    x1 = w00*n00 + w01*n01
    x2 = w02*n00 + w03*n01
    N10 = 1/(1+math.exp(-x1))
    N11 = 1/(1+math.exp(-x2))
    x3 = w10*N10 + w11*N11
    N20 = 1/(1+math.exp(-x3))
    # backpropagation to the output layer
    N20Ref = 1 if n00==1 and n01==1 else 0
    N20E = N20*(1-N20)*(N20Ref-N20)
    dW10 = b*N20E*N10
    dW11 = b*N20E*N11
    w10 = w10 + dW10 + a*pdW10
    w11 = w11 + dW11 + a*pdW11
    #pdW10 = w10 - w10_ant
    #pdW11 = w11 - w11_ant
    w10_ant = w10
    w11_ant = w11
    # backpropagation to the hidden layer
    N10E = N20E*w10
    N11E = N20E*w11
    dW00 = b*N10E*n00
    dW01 = b*N10E*n01
    dW02 = b*N11E*n00
    dW03 = b*N11E*n01
    w00 = w00 + dW00 + a*pdW00
    w01 = w01 + dW01 + a*pdW01
    w02 = w02 + dW02 + a*pdW02
    w03 = w03 + dW03 + a*pdW03
    #pdW00 = w00 - w00_ant
    #pdW01 = w01 - w01_ant
    #pdW02 = w02 - w02_ant
    #pdW03 = w03 - w03_ant
    w00_ant = w00
    w01_ant = w01
    w02_ant = w02
    w03_ant = w03

    
    n00 = 0
    n01 = 1
    # forward propagation
    x1 = w00*n00 + w01*n01
    x2 = w02*n00 + w03*n01
    N10 = 1/(1+math.exp(-x1))
    N11 = 1/(1+math.exp(-x2))
    x3 = w10*N10 + w11*N11
    N20 = 1/(1+math.exp(-x3))
    # backpropagation to the output layer
    N20Ref = 1 if n00==1 and n01==1 else 0
    N20E = N20*(1-N20)*(N20Ref-N20)
    dW10 = b*N20E*N10
    dW11 = b*N20E*N11
    w10 = w10 + dW10 + a*pdW10
    w11 = w11 + dW11 + a*pdW11
    #pdW10 = w10 - w10_ant
    #pdW11 = w11 - w11_ant
    w10_ant = w10
    w11_ant = w11
    # backpropagation to the hidden layer
    N10E = N20E*w10
    N11E = N20E*w11
    dW00 = b*N10E*n00
    dW01 = b*N10E*n01
    dW02 = b*N11E*n00
    dW03 = b*N11E*n01
    w00 = w00 + dW00 + a*pdW00
    w01 = w01 + dW01 + a*pdW01
    w02 = w02 + dW02 + a*pdW02
    w03 = w03 + dW03 + a*pdW03
    #pdW00 = w00 - w00_ant
    #pdW01 = w01 - w01_ant
    #pdW02 = w02 - w02_ant
    #pdW03 = w03 - w03_ant
    w00_ant = w00
    w01_ant = w01
    w02_ant = w02
    w03_ant = w03
    
    
    n00 = 0
    n01 = 0
    # forward propagation
    x1 = w00*n00 + w01*n01
    x2 = w02*n00 + w03*n01
    N10 = 1/(1+math.exp(-x1))
    N11 = 1/(1+math.exp(-x2))
    x3 = w10*N10 + w11*N11
    N20 = 1/(1+math.exp(-x3))
    # backpropagation to the output layer
    N20Ref = 1 if n00==1 and n01==1 else 0
    N20E = N20*(1-N20)*(N20Ref-N20)
    dW10 = b*N20E*N10
    dW11 = b*N20E*N11
    w10 = w10 + dW10 + a*pdW10
    w11 = w11 + dW11 + a*pdW11
    pdW10 = w10 - w10_ant
    pdW11 = w11 - w11_ant
    w10_ant = w10
    w11_ant = w11
    # backpropagation to the hidden layer
    N10E = N20E*w10
    N11E = N20E*w11
    dW00 = b*N10E*n00
    dW01 = b*N10E*n01
    dW02 = b*N11E*n00
    dW03 = b*N11E*n01
    w00 = w00 + dW00 + a*pdW00
    w01 = w01 + dW01 + a*pdW01
    w02 = w02 + dW02 + a*pdW02
    w03 = w03 + dW03 + a*pdW03
    pdW00 = w00 - w00_ant
    pdW01 = w01 - w01_ant
    pdW02 = w02 - w02_ant
    pdW03 = w03 - w03_ant
    w00_ant = w00
    w01_ant = w01
    w02_ant = w02
    w03_ant = w03
    
print("Error: "+str(N20E)+"\n\nPesos:")
print("W00: "+str(w00))
print("W01: "+str(w01))
print("W02: "+str(w02))
print("W03: "+str(w03))
print("W10: "+str(w10))
print("W11: "+str(w11))
```
